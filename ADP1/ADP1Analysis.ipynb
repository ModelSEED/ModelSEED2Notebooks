# Running Biolog in both models

```python
%run adp1analysisutil.py
ms2_model = util.msrecon.get_model("179225/Abaylyi_ADP1_RASTMS2_OMEGGA_Abaylyi_ADP1_RAST.mdlMS2_OMEGGA_iAbaylyi_Carbon_Succinic.gf")
published_model = MSModelUtil.from_cobrapy_json("data/iAbaylyi_ADP1.json")
ms2_output = util.simulate_biolog_phenotypes(
    ms2_model,
    objective="bio1",
    phenotypeset="157564/Full_Biolog_Dataset",
    base_uptake=0,
    global_atom_limits=None,
    base_media=None,
    ignore_experimental_data=True)
published_output = util.simulate_biolog_phenotypes(
    published_model,
    objective="bio1",
    base_media=None,
    phenotypeset="157564/Full_Biolog_Dataset",
    base_uptake=0,
    global_atom_limits=None,
    ignore_experimental_data=True)
ms2_output["details"].to_csv(util.output_dir+"/MS2-details.csv")
ms2_output["summary"].to_csv(util.output_dir+"/MS2-summary.csv")
published_output["details"].to_csv(util.output_dir+"/Published-details.csv")
published_output["summary"].to_csv(util.output_dir+"/Published-summary.csv")
```
